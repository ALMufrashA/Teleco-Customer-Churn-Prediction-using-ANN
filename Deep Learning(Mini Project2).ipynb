```python
#  To convert TotalCharges to numeric, fix missing
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)
data
```

# Performing Label Encoding

```python
# To change the string data into numercical data

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
# Encode categorical features
Text_Data = ['gender', 'Partner', 'Dependents', 'PhoneService', 'InternetService', 'Churn']
for col in Text_Data:
    data[col] = le.fit_transform(data[col])
le
```

# Scaling the numerical Columns

```python
# To change the gaps / values in the column of dataset
# Here to change the values of Age and income between 0 to 1 using "Scaler"

from sklearn.preprocessing import MinMaxScaler

# creating an object for scaling

scaler = MinMaxScaler()
```

```python
Num_Data = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']
#data[Num_Data] = scaler.fit_transform(data[Num_Data])
data[Num_Data] = scaler.fit_transform(data[Num_Data])
```

```python
data
```

```python
# To split the data into x and y

x = data[['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
          'PhoneService', 'InternetService', 'MonthlyCharges', 'TotalCharges']]
x
```

```python
y = data['Churn']  # 0 = No, 1 = Yes
y
```

# Splitting the data into X and Y for Training and Testing

```python
# To split the x and y data into training data  and testing data for a machine training purpose 

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size = 0.3)
```

```python
# To Compute the class weights for the data imbalance, we have to use this
# import the "class_weight" from the packages "utils"

from sklearn.utils import class_weight
```

```python
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(zip(np.unique(y_train), class_weights))
```

# ANN Model Building

```python
# To build neural network from the library "TensorFlow"

import tensorflow as tf

# Import keras from the package "tensorflow"

from tensorflow import keras
```

```python
# Build the neural network model

Model = keras.Sequential([
    keras.layers.Dense(32, input_shape = (x_train.shape[1],), activation = 'relu'),
    keras.layers.Dense(16, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

Model.compile(
    optimizer ='adam',
    loss ='binary_crossentropy',
    metrics = ['accuracy']
)

# Train the model
Model.fit(
    x_train, y_train,
    epochs = 20,
    batch_size = 32,
    validation_split = 0.2,
    class_weight = class_weights
)
```

```python
# To view the accuracy score of the model training.
Model.evaluate(x_test, y_test)
```
